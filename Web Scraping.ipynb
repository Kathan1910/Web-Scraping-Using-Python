```python
from autoscraper import AutoScraper

# URL to scrape
web_url = "https://www.offer18.com/awards"

# Provide example data that closely matches the webpage content
example_data = [
    "Awards 2023",
    "SPRING",
    "Offer18 has been recognized as both a Momentum Leader and High Performer in the G2 Spring Awards, while also being praised for being the easiest to use among its competitors."
]

# Create the scraper
scraper = AutoScraper()
scraper.build(web_url, example_data)

# Get the results
results = scraper.get_result_similar(web_url)

# Print the results
for result in results:
    print("Award:", result[0])
    print("Season:", result[1])
    print("Description:", result[2])
    print()
```

```python
from bs4 import BeautifulSoup
import requests

url = 'https://www.offer18.com/awards'

try:
    response = requests.get(url)
    response.raise_for_status()

    soup = BeautifulSoup(response.content, 'html.parser')

    # Find all elements containing awards and achievements information
    awards_elements = soup.select('div.text-3')

    for element in awards_elements:
        awards_text = element.get_text(strip=True)
        print("Awards & Achievements:")
        print(awards_text)
        print('-' * 40)

except Exception as e:
    print("An error occurred:", e)
```

```python
from bs4 import BeautifulSoup
import requests

url = 'https://www.offer18.com/awards'

try:
    response = requests.get(url)
    response.raise_for_status()

    soup = BeautifulSoup(response.content, 'html.parser')

    # Find all elements with class "award-wrapper-<year>"
    award_containers = soup.find_all(class_=lambda x: x and x.startswith("award-wrapper"))

    for container in award_containers:
        year = container.find_previous("h2").text.strip()  # Extract year from the preceding h2 element
        awards = container.find_all(class_="text-3")  # Find all awards within the container

        print(f"Awards {year}:")
        for award in awards:
            award_text = award.get_text(strip=True)
            print(f"- {award_text}")

        print('-' * 40)

except Exception as e:
    print("An error occurred:", e)
```

```python
from lxml import html
import requests

url = 'https://radixweb.com/our-awards'

try:
    response = requests.get(url)
    response.raise_for_status()

    tree = html.fromstring(response.content)

    # Extract the text of the h2 element
    heading = tree.xpath('/html/body/div[1]/div/div[13]/div/div/div[1]/div[1]/h2')[0].text.strip()

    # Find all div elements within the specified div using XPath
    award_divs = tree.xpath('/html/body/div[1]/div/div[13]/div/div/div[1]/div[1]/div')

    print(f"{heading}:")
    for div in award_divs:
        award_text = div.text_content().strip()
        print(f"- {award_text}")

except Exception as e:
    print("An error occurred:", e)
```

```python
from bs4 import BeautifulSoup
import requests

# List of website URLs
website_urls = ["https://www.devitpl.com/company/awards/", 
                "https://www.offer18.com/awards",]

# Initialize a list to store extracted data
scraped_data = []

for url in website_urls:
    # Fetch HTML content
    response = requests.get(url)
    html_content = response.content
    
    # Parse HTML using BeautifulSoup
    soup = BeautifulSoup(html_content, "html.parser")
    
    # Extract service details
    award_details = soup.find_all("div", class_="text-3")
    
    # Store data in a dictionary and append to the list
    data = {"url": url, "award": award_details}
    scraped_data.append(data)
```
