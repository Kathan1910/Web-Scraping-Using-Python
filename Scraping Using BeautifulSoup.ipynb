```python
## Modeling 

import requests
from bs4 import BeautifulSoup

# Define the base URL of the webpage you want to scrape
base_url = "https://hypeauditor.com/top-instagram-modeling-germany/?p="

# Initialize a counter for indexing
index = 1

# Initialize the starting page number
page_number = 1

while True:
    # Construct the URL for the current page
    url = f"{base_url}{page_number}"

    # Send an HTTP GET request to the URL to fetch the HTML content
    response = requests.get(url)

    # Check if the request was successful (status code 200)
    if response.status_code == 200:
        # Parse the HTML content using BeautifulSoup
        soup = BeautifulSoup(response.text, 'html.parser')

        # Find all elements that contain user details
        user_elements = soup.find_all(class_="contributor__name-content")

        # Check if there are no user elements on the page, indicating the end of the pages
        if not user_elements:
            break

        # Loop through the user elements and extract data for each user
        for user_element in user_elements:
            username = user_element.text.strip()
            follower_count = user_element.find_next(class_="row-cell subscribers").text.strip()
            country = user_element.find_next(class_="row-cell audience").text.strip()

            # Print the scraped data in the desired format with the index
            print(f"Index {index} - Username: @{username}, Follower Count: {follower_count}, Country: {country}")

            # Increment the index counter
            index += 1

        # Increment the page number to move to the next page
        page_number += 1
    else:
        print(f"Failed to retrieve page {page_number}. Status code: {response.status_code}")
        break
```

## Scrape All Three Togather And Export Into Csv Formate

```python
import requests
from bs4 import BeautifulSoup
import csv

# Function to scrape data from a given URL and return the updated index
def scrape_data(url, data, index):
    page_number = 1

    while True:
        current_url = f"{url}{page_number}"

        response = requests.get(current_url)

        if response.status_code == 200:
            soup = BeautifulSoup(response.text, 'html.parser')
            user_elements = soup.find_all(class_="contributor__name-content")

            if not user_elements:
                break

            for user_element in user_elements:
                username = user_element.text.strip()
                follower_count = user_element.find_next(class_="row-cell subscribers").text.strip()
                country = user_element.find_next(class_="row-cell audience").text.strip()

                data.append([f"Index {index}", f"@{username}", f"Follower Count: {follower_count}", f"Country: {country}"])
                index += 1

            page_number += 1
        else:
            print(f"Failed to retrieve page {page_number} from {url}. Status code: {response.status_code}")
            break

    return index  # Return the updated index

# Define the base URLs for the three categories
base_urls = [
    "https://hypeauditor.com/top-instagram-fashion-germany/?p=",
    "https://hypeauditor.com/top-instagram-beauty-germany/?p=",
    "https://hypeauditor.com/top-instagram-modeling-germany/?p="
]

# Initialize a list to store the scraped data
all_data = []

# Initialize an index counter
index = 1

# Scrape data from each URL and update the index
for url in base_urls:
    index = scrape_data(url, all_data, index)

# Define the CSV file name
csv_file = "scraped_data.csv"

# Write the data to a CSV file
with open(csv_file, 'w', newline='') as csvfile:
    csv_writer = csv.writer(csvfile)
    csv_writer.writerow(["Index", "Username", "Follower Count", "Country"])  # Write header row
    csv_writer.writerows(all_data)  # Write data rows

print(f"Scraped data from all URLs has been exported to {csv_file}.")
```

```python
import pandas as pd

# Read the CSV file into a Pandas DataFrame
csv_file = "scraped_data.csv"  # Replace with the actual CSV file name
df = pd.read_csv(csv_file)

# Extract the country names by removing the "Country: " prefix
df['Country'] = df['Country'].str.replace('Country: ', '')

# Filter the data to include only records with "Germany" as the country
germany_data = df[df['Country'] == 'Germany']

# Print the filtered data
print("Filtered Data with Germany as the Country:")
print(germany_data)

# If you want to save the filtered data to a new CSV file, you can use the following code:
germany_data.to_csv("germany_filtered_data.csv", index=False)
```
