```python
import requests
from bs4 import BeautifulSoup

# Define the URL of the Amazon product page
url = 'https://www.amazon.com/s?k=laptop&crid=2EWC6WOXXLQ7N&sprefix=lapto%2Caps%2C354&ref=nb_sb_noss_2'  # Replace with the product URL of your choice

# Send an HTTP GET request to the URL
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    # Parse the HTML content using BeautifulSoup
    soup = BeautifulSoup(response.text, 'html.parser')

    # Extract the number of customer reviews
    review_count_elem = soup.find('span', {'id': 'acrCustomerReviewText'})
    if review_count_elem:
        review_count_text = review_count_elem.text.strip()
        # Extract the number from the text (e.g., "4,567 ratings" becomes 4567)
        review_count = int(''.join(filter(str.isdigit, review_count_text)))
    else:
        review_count = 'N/A'
        
    # Extract the average star rating
    star_rating_elem = soup.find('span', {'class': 'a-icon-alt'})
    if star_rating_elem:
        star_rating_text = star_rating_elem.text.strip()
        # Extract the numeric part of the rating (e.g., "4.5 out of 5 stars" becomes 4.5)
        star_rating = float(star_rating_text.split(' ')[0])
    else:
        star_rating = 'N/A'

    print(f'Number of Customer Reviews: {review_count}')
    print(f'Average Star Rating: {star_rating}')

else:
    print('Failed to retrieve the webpage.')
```
