```python
import pandas as pd

excel_file = "scraped_data.xlsx" # Replace with the path to your Excel file
df = pd.read_excel(excel_file)

# Replace 'Email Column Name' with the actual column name containing emails
duplicates = df[df['Company Email'].duplicated(keep=False)]
duplicate_count = len(duplicates)

unique_emails = df[~df['Company Email'].duplicated(keep=False)]
unique_count = len(unique_emails)

print(f"Duplicate Email Count: {duplicate_count}")
print(f"Unique Email Count: {unique_count}")
```

```python
import pandas as pd

excel_file = "scraped_data.xlsx"  # Replace with the path to your Excel file
df = pd.read_excel(excel_file)

# Replace 'Email Column Name' with the actual column name containing emails
unique_emails = df[~df['Company Email'].duplicated(keep=False)]

# Specify the path and name for the new Excel file
output_excel_file = 'unique_emails.xlsx'

# Save the DataFrame with unique emails to a new Excel file
unique_emails.to_excel(output_excel_file, index=False)
```
