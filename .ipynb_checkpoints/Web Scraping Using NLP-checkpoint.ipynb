```python
import requests
import spacy
from bs4 import BeautifulSoup

# Load the SpaCy model
nlp = spacy.load("en_core_web_sm")

# Function to extract relevant text from HTML elements using CSS selectors
def extract_relevant_text(soup, selectors):
    relevant_text = []
    for selector in selectors:
        elements = soup.select(selector)
        for element in elements:
            relevant_text.append(element.get_text())
    return " ".join(relevant_text)

# Define key phrases and associated keywords
key_phrases = {
    "achievements": ["achieved", "success", "accomplishment"],
    "awards": ["award", "recognized", "honored"],
    "testimonials": ["testimonial", "feedback", "review"],
    "clients": ["client", "customer", "partner"]
    # Add more key phrases and keywords as needed
}

# Function to summarize key takeaways from text with a specified maximum length
def summarize_key_takeaways(text, max_characters=500):
    key_takeaways = []
    current_summary = []
    characters_count = 0

    for sent in nlp(text).sents:
        if len(sent.text.strip()) > 0:
            current_summary.append(sent.text.strip())
            characters_count += len(sent.text)
            if characters_count >= max_characters:
                key_takeaways.append(" ".join(current_summary))
                break

    return key_takeaways

# URL of the website you want to scrape
url = "https://thriveagency.com/"
# Send an HTTP request to fetch the HTML content
response = requests.get(url)
html_content = response.content

# Create a BeautifulSoup object to parse HTML content
soup = BeautifulSoup(html_content, "html.parser")

# Define CSS selectors for relevant content
selectors = [
    "h1", "h2", "h3", "h4", "h5", "h6", "p", "li", "div", "span", "a"
]

# Extract relevant text using specified CSS selectors
relevant_text = extract_relevant_text(soup, selectors)

# Extract information based on key phrases and keywords
extracted_info = {}

for phrase, keywords in key_phrases.items():
    relevant_sections = []
    for sent in nlp(relevant_text).sents:
        if any(keyword in sent.text.lower() for keyword in keywords):
            relevant_sections.append(sent.text.strip())
    if relevant_sections:
        extracted_info[phrase] = relevant_sections

# Summarize extracted information with bullet points and limit to 2 or 3 sentences
summarized_info = {}

for phrase, sections in extracted_info.items():
    # Remove duplicate sections
    unique_sections = list(set(sections))

    # Summarize to a maximum of 2 or 3 sentences and 500 characters
    summarized_text = "\n\n".join(summarize_key_takeaways("\n".join(unique_sections), max_characters=500))

    summarized_info[phrase] = summarized_text

# Print the summarized information
for phrase, summary in summarized_info.items():
    print(f"{phrase.capitalize()}:\n{summary.strip()}\n")
```

## Transfer Learning 

```python
import requests
from bs4 import BeautifulSoup
from transformers import GPT2Tokenizer, GPT2LMHeadModel

# Function to scrape text from a URL
def scrape_text(url):
    try:
        response = requests.get(url)
        soup = BeautifulSoup(response.text, "html.parser")
        paragraphs = soup.find_all("p")
        text = " ".join([p.get_text() for p in paragraphs])
        return text
    except Exception as e:
        print(f"Error scraping {url}: {str(e)}")
        return ""

# Function to summarize text using GPT-2 model
def summarize_text(text):
    try:
        model_name = "gpt2"  # You can use other GPT-2 variants
        model = GPT2LMHeadModel.from_pretrained(model_name)
        tokenizer = GPT2Tokenizer.from_pretrained(model_name)
        
        input_ids = tokenizer.encode(text, return_tensors="pt", max_length=1024, truncation=True)
        summary_ids = model.generate(input_ids, max_length=150, num_return_sequences=1, no_repeat_ngram_size=2, top_k=50, top_p=0.95)
        
        summary = tokenizer.decode(summary_ids[0], skip_special_tokens=True)
        return summary
    except Exception as e:
        print(f"Error summarizing text: {str(e)}")
        return ""

# List of URLs to scrape
urls = [
    "https://www.hexacta.com/",
    
]

# Scrape and summarize text for each URL
for url in urls:
    scraped_text = scrape_text(url)
    if scraped_text:
        summary = summarize_text(scraped_text)
        print("URL:", url)
        print("Summary:", summary)
        print("=" * 50)
```

```python
import requests
import spacy
from bs4 import BeautifulSoup
from transformers import GPT2Tokenizer, GPT2LMHeadModel

# Load the SpaCy model
nlp = spacy.load("en_core_web_sm")

# Load GPT-2 model and tokenizer
model_name = "gpt2"  # You can use other GPT-2 variants
model = GPT2LMHeadModel.from_pretrained(model_name)
tokenizer = GPT2Tokenizer.from_pretrained(model_name)

# Function to extract relevant text from different HTML tags
def extract_relevant_text(soup, relevant_tags):
    relevant_text = []
    for tag in soup.find_all(relevant_tags):
        relevant_text.append(tag.get_text())
    return " ".join(relevant_text)

# Function to summarize text using GPT-2 model
def summarize_text(text):
    input_ids = tokenizer.encode(text, return_tensors="pt", max_length=1024, truncation=True)
    summary_ids = model.generate(input_ids, max_length=150, num_return_sequences=1, no_repeat_ngram_size=2, top_k=50, top_p=0.95)
    summary = tokenizer.decode(summary_ids[0], skip_special_tokens=True)
    return summary

# Define key phrases and associated keywords
key_phrases = {
    "achievements": ["achieved", "success", "accomplishment"],
    "awards": ["award", "recognized", "honored"],
    "testimonials": ["testimonial", "feedback", "review"],
    # Add more key phrases and keywords
}

# Function to categorize content based on key phrases
def categorize_content(text, key_phrases):
    categorized_content = {phrase: [] for phrase in key_phrases}

    for phrase, keywords in key_phrases.items():
        for line in text.split('\n'):
            if any(keyword in line.lower() for keyword in keywords):
                categorized_content[phrase].append(line.strip())

    return categorized_content

# Function to summarize key takeaways from text using GPT-2
def summarize_key_takeaways(text, max_lines=2):
    return summarize_text(text)

# Function to remove duplicates from a list while preserving the order
def remove_duplicates(seq):
    seen = set()
    seen_add = seen.add
    return [x for x in seq if not (x in seen or seen_add(x))]

# URL of the website you want to scrape
url = "https://thriveagency.com/"  # Replace with the URL of the website you want to scrape

# Send an HTTP request to fetch the HTML content
response = requests.get(url)
html_content = response.content

# Create a BeautifulSoup object to parse HTML content
soup = BeautifulSoup(html_content, "html.parser")

# Extract relevant text using tags that often contain content of interest
relevant_tags = ["h1", "h2", "h3", "h4", "h5", "h6", "p", "li"]
relevant_text = extract_relevant_text(soup, relevant_tags)

# Categorize the content
categorized_content = categorize_content(relevant_text, key_phrases)

# Remove duplicates from categorized content
for section, content in categorized_content.items():
    if content:
        categorized_content[section] = remove_duplicates(content)

# Print or format the categorized content
for section, content in categorized_content.items():
    if content:
        print(f"{section.capitalize()}:")
        for item in content:
            print(f"- {item}")
```

## Experiment With JavaScripted WebUrls

```python
import requests
import spacy
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.chrome.service import Service as ChromeService
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options as ChromeOptions
from selenium.webdriver.firefox.service import Service as FirefoxService
from selenium.webdriver.firefox.options import Options as FirefoxOptions

# Load the SpaCy model
nlp = spacy.load("en_core_web_sm")

# Function to extract relevant text from different HTML tags
def extract_relevant_text(soup, relevant_tags):
    relevant_text = []
    for tag in soup.find_all(relevant_tags):
        relevant_text.append(tag.get_text())
    return " ".join(relevant_text)

# Define key phrases and associated keywords
key_phrases = {
    "achievements": ["achieved", "success", "accomplishment"],
    "awards": ["award", "recognized", "honored"],
    "testimonials": ["testimonial", "feedback", "review"],
    # Add more key phrases and keywords
}

# Function to summarize key takeaways from text
def summarize_key_takeaways(text, max_lines=2):
    key_takeaways = []
    sentences = text.split(".")
    current_lines = 0
    current_summary = []

    for sentence in sentences:
        if len(sentence.strip()) > 0:
            current_summary.append(sentence.strip() + ".")
            current_lines += 1

        if current_lines >= max_lines:
            key_takeaways.append(" ".join(current_summary))
            current_summary = []
            current_lines = 0

    if current_summary:
        key_takeaways.append(" ".join(current_summary))

    return key_takeaways

# URL of the website you want to scrape
# url = "https://www.hexacta.com/"
url ="https://www.awardforce.com/"
# Set up a Selenium WebDriver to render the page
chrome_options = ChromeOptions()
chrome_options.add_argument("--headless")
chrome_service = ChromeService("E:\Web Scraping\chromedriver-win64\chromedriver.exe")  # Replace with the actual path to chromedriver executable
driver = webdriver.Chrome(service=chrome_service, options=chrome_options)
driver.get(url)

# Extract the HTML content and JavaScript code
html_content = driver.page_source
javascript_code = driver.execute_script("return document.documentElement.outerHTML")

# Create a BeautifulSoup object to parse HTML content
soup = BeautifulSoup(html_content, "html.parser")

# Extract relevant text using tags that often contain content of interest
relevant_tags = ["h1", "h2", "h3", "h4", "h5", "h6", "p", "li"]
relevant_text = extract_relevant_text(soup, relevant_tags)

# Extract information based on key phrases and keywords
extracted_info = {}

for phrase, keywords in key_phrases.items():
    relevant_sections = []
    for sent in nlp(relevant_text).sents:
        if any(keyword in sent.text.lower() for keyword in keywords):
            relevant_sections.append(sent.text.strip())
    if relevant_sections:
        extracted_info[phrase] = relevant_sections

# Summarize extracted information
summarized_info = {}

for phrase, sections in extracted_info.items():
    # Remove duplicate sections
    unique_sections = list(set(sections))
    
    # Summarize using the first 2 sentences from each unique section
    summarized_sections = []
    for section in unique_sections:
        sentences = section.split(".")
        if len(sentences) >= 2:
            summarized_sections.append(sentences[0] + "." + sentences[1])
        else:
            summarized_sections.append(section)
    
    summarized_info[phrase] = summarized_sections

# Create a dictionary to store summaries by topic
topic_summaries = {phrase: [] for phrase in key_phrases}

# Populate the topic_summaries dictionary with two-line summaries for each topic
for phrase, sections in summarized_info.items():
    if phrase in topic_summaries:
        topic_summary_lines = []
        for section in sections:
            sentences = section.split(".")
            if len(sentences) >= 2:
                topic_summary_lines.extend(sentences[:2])
            else:
                topic_summary_lines.extend(sentences)
        topic_summary = " ".join(topic_summary_lines)
        topic_summaries[phrase].append(topic_summary)

# Print the topic-wise summaries with two lines each
for phrase, summaries in topic_summaries.items():
    if summaries:
        print(f"{phrase.capitalize()}:")
        for summary in summaries:
            print("- " + summary)
        print()

# Close the Selenium WebDriver
driver.quit()
```

```python
import requests
from bs4 import BeautifulSoup

# URL of the website you want to scrape
url = "https://www.awardforce.com/"

# Send an HTTP request to fetch the HTML content
response = requests.get(url)

# Check if the request was successful (status code 200)
if response.status_code == 200:
    # Parse the HTML content
    html_content = response.text

    # Create a BeautifulSoup object to parse HTML content
    soup = BeautifulSoup(html_content, "html.parser")

    # Extract relevant text using tags that often contain content of interest
    relevant_tags = ["h1", "h2", "h3", "h4", "h5", "h6", "p", "li"]
    relevant_text = []
    for tag in soup.find_all(relevant_tags):
        relevant_text.append(tag.get_text())

    # Combine the relevant text into a single string
    extracted_text = " ".join(relevant_text)

    # Print the extracted content
    print(extracted_text)

else:
    print("Failed to retrieve the web page. Status code:", response.status_code)
```
