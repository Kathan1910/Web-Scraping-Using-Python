```python
from bs4 import BeautifulSoup
import requests

# Provide the URL of the webpage
url = 'https://www.yellowpages.com/search?search_terms=software+development+companies&geo_location_terms=new+york/'

# Send an HTTP GET request to the URL
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    html_content = response.text
    soup = BeautifulSoup(html_content, 'html.parser')

    # Find all elements that contain company information
    company_elements = soup.find_all('div', class_='info')

    # Loop through the company elements and extract details
    for company_element in company_elements:
        company_name_element = company_element.find('span')
        if company_name_element:
            company_name = company_name_element.text.strip()

        phone_number_element = company_element.find('div', class_='phones')
        if phone_number_element:
            phone_number = phone_number_element.text.strip()

        website_element = company_element.find('a', class_='track-visit-website')
        if website_element:
            website_url = website_element.get('href')

        city_element = company_element.find('div', class_='locality')
        if city_element:
            city = city_element.text.strip()

        print(f'Company Name: {company_name}')
        print(f'Phone Number: {phone_number}')
        print(f'Website: {website_url}')
        print(f'City: {city}')
        print()  # Add a blank line for separation

else:
    print('Failed to retrieve the web page')
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time

# Set up the web driver without specifying executable_path
driver = webdriver.Chrome()

# Provide the URL of the webpage
url = 'https://www.yellowpages.com/search?search_terms=software+development+companies&geo_location_terms=new+york/'

# Open the webpage
driver.get(url)

# Create an empty list to store the URLs
company_urls = []

while True:
    # Find all elements that contain company information on the current page
    company_elements = driver.find_elements(By.CLASS_NAME, 'info')

    for company_element in company_elements:
        # Extract the URL from the "a" tag within the company element
        company_url_element = company_element.find_element(By.TAG_NAME, 'a')
        company_url = company_url_element.get_attribute('href')
        company_urls.append(company_url)

    # Check if there is a "Next" button to navigate to the next page
    next_button = driver.find_element(By.CLASS_NAME, 'next')
    if next_button and 'disabled' not in next_button.get_attribute('class'):
        next_button.click()
        time.sleep(2)  # Adding a small delay to ensure the next page loads
    else:
        break  # Break the loop if there is no more "Next" button

# Print the collected company URLs
for url in company_urls:
    print(url)

# Close the web driver when done
driver.quit()
```

## Scrape The Data From The URLs

```python
import requests
from bs4 import BeautifulSoup
import csv

# Define a function to extract the required information from a URL
def scrape_url(url):
    try:
        response = requests.get(url)
        soup = BeautifulSoup(response.text, 'html.parser')

        # Extract company name
        company_name_element = soup.find('h1', class_='dockable business-name')
        company_name = company_name_element.text.strip() if company_name_element else 'Not available'

        # Extract company website
        company_website_element = soup.find('a', class_='website-link')
        company_website = company_website_element['href'].strip() if company_website_element else 'Not available'

        # Extract company email
        company_email_element = soup.find('a', class_='email-business')
        company_email = company_email_element['href'].strip() if company_email_element else 'Not available'

        # Extract company address
        company_address_element = soup.find('span', class_='address')
        company_address = company_address_element.text.strip() if company_address_element else 'Not available'

        return {
            'Company Name': company_name,
            'Company Website': company_website,
            'Company Email': company_email,
            'Company Address': company_address
        }
    except Exception as e:
        return {
            'Error': str(e)
        }

# Read the list of URLs from a text file
with open('urls1.txt', 'r') as file:
    urls = file.read().splitlines()

# Create a list to store the scraped data
scraped_data = []

# Loop through each URL and scrape the information
for url in urls:
    scraped_data.append(scrape_url(url))

# Define the CSV file path
csv_file = 'scraped_data1.csv'

# Write the scraped data to a CSV file
with open(csv_file, 'w', newline='', encoding='utf-8') as csvfile:
    fieldnames = ['Company Name', 'Company Website', 'Company Email', 'Company Address']
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    
    # Write the header row
    writer.writeheader()
    
    # Write the data rows
    for data in scraped_data:
        writer.writerow(data)

print(f'Scraped data has been saved to {csv_file}')
```
